#1. Cargar los archivos

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivos
df_old = pd.read_excel("/content/drive/MyDrive/TFG/205865-2-intensidad-media-laborables 2021-2023-xls.xls")
df_new = pd.read_excel("/content/drive/MyDrive/TFG/205865-4-intensidad-media-laborables 2024-2025-xlsx.xlsx")
```

#2. Columnas

```python
# Ver columnas
print(df_old.columns)
print(df_new.columns)
```

#3. Columnas del interior de la M-30

```python
# Columnas del interior de la M-30
cols_m30 = [
    'Interior 1er  cinturón',
    'En el 1er  cinturón',
    'Entre 1er y 2º cinturón',
    'En el 2er  cinturón',
    'Entre  2º cinturón y M-30',
    'M-30'
]
```

#4. Filtrar sola 2021 y 2025

```python
# Filtrar solo 2021 y 2025
df_2021 = df_old[df_old['Año'] == 2021].copy()
df_2025 = df_new[df_new['Año'] == 2025].copy()
```

#5. Calcular la intensidad media del interior de la M-30

```python
# Calcular intensidad media del área interior de la M-30
df_2021['Media_M30_interior'] = df_2021[cols_m30].mean(axis=1)
df_2025['Media_M30_interior'] = df_2025[cols_m30].mean(axis=1)
```

#6. Media anual

```python
# Media anual
media_2021 = df_2021['Media_M30_interior'].mean()
media_2025 = df_2025['Media_M30_interior'].mean()

print("Media interior M-30 en 2021:", round(media_2021, 2))
print("Media interior M-30 en 2025:", round(media_2025, 2))
```

#7. Calcular variación %

```python
# Variación %
variacion = ((media_2025 - media_2021) / media_2021) * 100
print("Variación porcentual 2021-2025:", round(variacion, 2), "%")
```

#8. Gráfico

```python
# Gráfico
años = ['2021', '2025']
valores = [media_2021, media_2025]

plt.figure(figsize=(7,5))
plt.bar(años, valores)
plt.title("Comparación de intensidad media de tráfico\nInterior de la M-30")
plt.ylabel("Intensidad media")

# Añadir los valores en la parte superior de cada barra
for i, valor in enumerate(valores):
    plt.text(i, valor, str(round(valor, 2)), ha='center', va='bottom')

plt.show()
```

El tráfico presenta un ligero incremento (+2,46%), manteniéndose en niveles elevados en el interior de la M-30, lo que refleja una situación estructural de alta congestión

Aunque el incremento porcentual de la intensidad media del tráfico entre 2021 y 2025 es moderado (2,46%), en términos absolutos representa un aumento significativo del volumen de circulación en una de las zonas más congestionadas de la ciudad. Esto evidencia que la presión sobre la red viaria y el sistema de estacionamiento continúa siendo elevada
